# 暂态现象
## 已知条件
单相无限大功充率电流供电系统，如 [教材] [1] ![图1](FIG8-1.jpg)所示



## 分析过程
电路的的瞬时电压平衡方程为：

$$Ri(t)+L\frac{di(t)}{dt}=V_m\sin(\omega t+\alpha)$$

解出瞬时电流为：

$$i(t)=I_m\sin(\omega t+\alpha-\theta)-I_me^{-\frac{t}{\tau}}\sin(\alpha-\theta)$$
其中

电流幅值$$I_m=V_m/Z$$

时间常数$$\tau=L/R$$

阻抗角$$\theta=\arctan{ \frac{\omega L}{R}}$$  

阻抗值$$Z=\sqrt{R^2+X^2}$$

观察瞬时电流的表达式可知，瞬时电流由两项构成。第一项是稳态正弦值，第二项为衰减直流分量。而且当$t=0$时，即扰动最初，两项大小相等符号相反，合成为值0。此时，由$\alpha$来确定暂态电流的大小。

因此，当$\alpha=\theta-\pi/2$时，则直流分量达到最大值$I_m=V_m/Z$与交流分量的峰值相等。

考虑到一般线路$\omega L \gg R$，则$\theta\approx \pi/2$，此时，在电路电压为最大值时闭合无直流分量；在电路电压为０时，直流分量反而最大。
## 仿真
设$R=0.125\Omega$,$L=10mH$,电压源电压$v(t)=151\sin(377t+\alpha)$,根据情况模拟其动态过程。

$$Z=0.125+j(377)(0.01)=3.772\angle 88.1^\circ$$
$$I_m=\frac{151}{3.772}=40A$$
$$\tau=\frac{L}{R}=0.08 sec$$
所以，瞬时电流表达式为：
$$i(t)=40\sin(\omega t+\alpha-88.1^\circ)-40e^{-t/0.08}\sin(\alpha-88.1^\circ)$$

故当$\alpha=88.1^\circ$时无直流分量；当$\alpha=88.1^\circ-90^\circ=-1.9^\circ$时，直流分量最大。这两种情况的模拟代码如下：


```python
%matplotlib inline
from math import sin,pi,exp
import numpy as np
import matplotlib.pyplot as plt
alf1=88.1*pi/180
theta=88.1*pi/180
t=np.arange(0,0.3,0.001)
i1=[]
for tp in t:
    i=40*sin(377*tp+alf1-theta)-40*exp(-tp/.08)*sin(alf1-theta)
    i1.append(i)
plt.plot(t,i1)
```

```python
%matplotlib inline
from math import sin,pi,exp
import numpy as np
import matplotlib.pyplot as plt
alf2=-1.9*pi/180
theta=88.1*pi/180
t=np.arange(0,0.3,0.001)
i1=[]
for tp in t:
    i=40*sin(377*tp+alf2-theta)-40*exp(-tp/.08)*sin(alf2-theta)
    i1.append(i)
plt.plot(t,i1)
```
